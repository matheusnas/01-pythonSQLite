```python
import duckdb
import pandas as pd
import os
from datetime import datetime
```

```python
connection = duckdb.connect(database='dados_duckdb.db', read_only=False)
```

```python
arquivo = "z0019_2.csv"
data_ingestão = datetime.now()
df = pd.read_csv(f"../landing/{arquivo}", sep=";")
df['arquivo'] = arquivo
df['data_ingestão'] = data_ingestão

df
```

```python
connection.execute("""

CREATE TABLE IF NOT EXISTS bronze_produto (
                   NATBR VARCHAR,
                   MAKTX VARCHAR,
                   WERKS VARCHAR,
                   MAINS VARCHAR,
                   LABST VARCHAR,     
                   arquivo VARCHAR,
                   data_ingestão TIMESTAMP
                )
""")
```

```python
connection.execute("INSERT INTO bronze_produto select * from df")
```

```python
result = connection.execute("SELECT * FROM bronze_produto").fetchdf()
result
```
